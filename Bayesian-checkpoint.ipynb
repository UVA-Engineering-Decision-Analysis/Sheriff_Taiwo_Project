**Probability of Matriculation and Optimal Financial Aid Allocation**
    
 Rising awareness of student debt and college affordability are currently at the forefront of many incoming college students and their parents. To combat out of control costs, public and private institutions are looking towards data driven methods to streamline marketing costs, create tuition pricing strategies to compete for students, drive enrollment, and increase revenue. The decision maker(s) are college recruiters, and financial appropriation teams.  

 College recruiters want to spend their time on  effort on marketing to admitted candidates that will likely matriculate while financial teams want to offer the optimal amount of financial aid required to keep admitted candidates with high likelihood of matriculation and nudge students with mid to low probabilities regions into matriculating. 
     
 This tool will help decision makers cut costs by using effective target marketing and only offer the optimal amount of financial aid dollars. Using two years of student data, classification models will be developed for first time incoming student (first time incoming student refers to student that have never been admitted to a four-year college). Three admitted student segments namely; In-State, Out of State and International student segments will be analyzed, to predict the likelihood of matriculation.
    
 The predictive tool will help college recruiters streamline their marketing efforts towards students with High Likelihood of Matriculation by assigning a probability to matriculate to each admitted student. The optimization portion of this tool aims to achieve two things; use probability results from the classification models to identify and market to students with high likelihood of matriculation and offer optimal financial aid designed for student at different probability ranges. The client and owner of this data has donated two years’ worth of data for research purposes. 

 The data has been thoroughly scrubbed of Personal Identifiable Information to prevent client and student info from ending up in public domain. Python and associated data science and pricing optimization libraries will be the tool of choice.
     
** Mathematical Formalism ** 
 The decision-maker's action set $A$: Based on the likelihood to matriculate probability scores assigned to each applicant from the classification algorithm, the decision-maker has to decide the following;

- In general, applicants with high probability scores (0.65 and above) tend to matriculate. Therefore, a flat dollar amount or no additional funding is offered to entice those candidates.  

- Applicants with mid probability scores (0.4 – 0.65) tend to have competing offers from other schools and as a result, a variable optimal dollar amount is offered to “win” them over. 

- Applicants with low probability scores (0.3 and below) tend to not matriculate in general. Therefore, a flat dollar amount or no additional funding is offered to entice those candidates.
 The $State  Space$ is characterized by three principles: {$S$,$A$,$A$ction(s),$R$esult(s,a),$C$ost(s,a)}, where: 
 - $S$ is the set of all variables per student leading to matriculation, in our case likelihood to matriculate probabilities
 
 - $A$ is the set of all possible actions regarding the states or probabilities 
 
- $A$ction(s) is the set of all possible actions a decision-maker can take based on the probabilities associated to each candidate. In our case, the decision-maker can choose to provide additional financial aid to a candidate in a bid to sway them towards matriculating to the school.
 
 
 - $R$esult(s,a) is the is the set of all possible results based on the action of the decision-maker. For the purpose of this exercise, we can envision the following results:
 
     * That the student in the high probability range (0.65 and above) will as predicted matriculate to the school and therefore, additional financial aid isn’t required.
 
     * That a sizeable population of students in the mid probability range (0.4 - 0.65) will indeed react positively to the additional funds being offered and matriculate to the school.
 
     * That a large population of the students in lower probability range (0.3 and below) won’t matriculate and therefore, offering additional isn’t necessary.
 
 
 - $C$ost(s,a) is the set of possible costs associated with the actions of the decision-make. For the purpose of the decision-maker, we can envision the following results:
 
      * That some student in the high probability range (0.65 and above) may not matriculate due to flat financial aid  package.#       
    
      * That some student in the mid probability range (0.4 and 0.65), specifically students in the lower band of this bucket may not matriculate due to the larger financial aid required to tip them into the high probability bucket.
       
      * That some students in the low probability range (0.3 and below), specifically students in the higher band of this bucket may indeed matriculate if larger financial aid was to tip them into middle band of the next probability buckets.

```python
def na_distance(myrow):
    return np.nan if pd.isnull(myrow['LATD_NBR']) | pd.isnull(myrow['LGTD_NBR']) else  \
        geopy.distance.distance((42.745357, -84.470605), (myrow['LATD_NBR'], myrow['LGTD_NBR'])).miles

#  creates distance to MSU synthetic feature
raw_df['distance2MSU'] = raw_df.apply(na_distance,axis=1)

# create days since may 1 synthetic feature
raw_df['days_since_may'] = abs(raw_df.APPLICATION_TERM.map(lambda x:
                                                           pd.datetime.strptime(
                                                               '05-01-' + x.replace('FS', ''),
                                                                '%m-%d-%y')) - raw_df.APPLICATION_DATE).map(lambda x: x.days)
raw_df['need'] = raw_df.COA - raw_df.EFC1
raw_df.loc[raw_df.need <= 0, 'need'] = 0
raw_df['FINANCIAL_AID_TOTAL'] = raw_df.FINANCIAL_AID_TOTAL - raw_df.MSU_ACPT_GIFT_AID  #FINANCIAL_AID Total excluding Igrant 
raw_df['Yield_Indicator'] = raw_df.MATRICULANTS

# Create Best Score Synthetic Feature
sat_bins = [0, 500, 550, 590, 620, 650, 690, 730, 780, 830, 880, 920, 960, 990, 1030, 1060, 1100, 1130, 1160, 1200,
            1230, 1260, 1300, 1330, 1360, 1390, 1420, 1450, 1490, 1530, 1570, 1601]
raw_df['SAT_CONV'] = pd.cut(raw_df.SAT_COMBINED, bins= sat_bins, right=False, labels=False) + 6
raw_df.loc[raw_df.SAT_CONV == 6, 'SAT_CONV'] = 5  # correct because lowest category is 5 not 6
raw_df['BEST_SCORE'] = raw_df[['SAT_CONV', 'ACT_COMPOSITE']].max(axis=1, skipna=True)  # take highest score no NA

# Create School Feeder Synthetic Feature
school_count = raw_df.groupby(  # create dataframe with school identifier and number of applicants
    'SCHOOL_CEEB').size().reset_index().rename(columns={0: 'school_feeder'})
raw_df = raw_df.merge(school_count, on='SCHOOL_CEEB', how = 'left')
```

```python
####### Data Imputation ###### 
def school_zip_impute(df, col):
    mean_df = pd.DataFrame(df.groupby('SCHOOL_ZIP_CODE')[col].mean())
    mean_df_state = pd.DataFrame(df.groupby('SCHOOL_STATE')[col].mean())
    mean_df_segment = pd.DataFrame(df.groupby('STUDENT_SEGMENT')[col].mean())
    df[col] = df[col].where(df[col].notnull(), df['SCHOOL_ZIP_CODE'].map(mean_df[col]))
    df[col] = df[col].where(df[col].notnull(), df['SCHOOL_STATE'].map(mean_df_state[col]))
    df[col] = df[col].where(df[col].notnull(), df['STUDENT_SEGMENT'].map(mean_df_segment[col]))
    return df

def school_impute(df, col):
    mean_df_ind = pd.DataFrame(df.groupby('SCHOOL_CEEB')[col].mean())
    mean_df = pd.DataFrame(df.groupby('SCHOOL_ZIP_CODE')[col].mean())
    mean_df_state = pd.DataFrame(df.groupby('SCHOOL_STATE')[col].mean())
    mean_df_segment = pd.DataFrame(df.groupby('STUDENT_SEGMENT')[col].mean())
    df[col] = df[col].where(df[col].notnull(), df['SCHOOL_CEEB'].map(mean_df_ind[col]))
    df[col] = df[col].where(df[col].notnull(), df['SCHOOL_ZIP_CODE'].map(mean_df[col]))
    df[col] = df[col].where(df[col].notnull(), df['SCHOOL_STATE'].map(mean_df_state[col]))
    df[col] = df[col].where(df[col].notnull(), df['STUDENT_SEGMENT'].map(mean_df_segment[col]))
    return df


def non_fafsa_zip(df, col):
    mean_df = pd.DataFrame(df[df['FAFSA'] == 'N'].groupby('SCHOOL_ZIP_CODE')[col].mean())
    mean_df_state = pd.DataFrame(df[df['FAFSA'] == 'N'].groupby('SCHOOL_STATE')[col].mean())
    mean_df_segment = pd.DataFrame(df[df['FAFSA'] == 'N'].groupby('STUDENT_SEGMENT')[col].mean())
    df[col] = df[col].where(df[col].notnull(), df['SCHOOL_ZIP_CODE'].map(mean_df[col]))
    df[col] = df[col].where(df[col].notnull(), df['SCHOOL_STATE'].map(mean_df_state[col]))
    df[col] = df[col].where(df[col].notnull(), df['STUDENT_SEGMENT'].map(mean_df_segment[col]))
    return df
```
